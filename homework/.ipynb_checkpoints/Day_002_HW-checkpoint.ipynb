```python
import os
import numpy as np
import pandas as pd
```

```python
# 設定 data_path
dir_data = './data/'
```

```python
f_app = os.path.join(dir_data, 'application_train.csv')
print('Path of read in data: %s' % (f_app))
app_train = pd.read_csv(f_app)
# print(type(app_train) : dataframe
```

### 如果沒有想法，可以先嘗試找出剛剛例子中提到的問題的答案
#### 資料的 row 數以及 column 數

```python
print("len of column : %d" %len(app_train.columns))
print("len of row : %d" %len(app_train.index))
```

#### 列出所有欄位

```python
app_train
```

#### 截取部分資料

```python
#抽出男性資料擁有車且總收入大於50W
df_M = app_train[(app_train.CODE_GENDER == "M") & (app_train.FLAG_OWN_CAR == "Y") & (app_train.AMT_INCOME_TOTAL>500000)]

print("符合人數 ： %d" %len(df_M.index))

df_M
```
